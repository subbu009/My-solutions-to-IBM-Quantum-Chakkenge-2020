```python
from qiskit import *
```

```python
address_qubits = QuantumRegister(4,name = 'add')
row_qubits = QuantumRegister(4,name = 'row')
column_qubits = QuantumRegister(4,name = 'column')
search_qubits = QuantumRegister(8,name = 'search')
counter_qubits = QuantumRegister(4,name = 'count')
flag_qubit = QuantumRegister(1,name = 'flag')
classical_bits = ClassicalRegister(4,name = 'c_bits')
qc = QuantumCircuit(address_qubits, row_qubits,column_qubits,search_qubits, counter_qubits, flag_qubit, classical_bits)
qc.h(address_qubits)
qc.x(flag_qubit)
qc.h(flag_qubit)
qc.barrier()
```

```python
def qram(qc,address,row,column,problem_set):
    for m in problem_set:
        if m == problem_set[0]:
            qc.x(address)
            for i in m:
                qc.mct([address[0],address[1],address[2],address[3],row[int(i[0])]],column[int(i[1])],mode = 'noancilla')
            qc.x(address)
            qc.barrier()
        if m == problem_set[1]:
            qc.x(address[0:3])
            for i in m:
                qc.mct([address[0],address[1],address[2],address[3],row[int(i[0])]],column[int(i[1])],mode = 'noancilla') 
            qc.x(address[0:3])
            qc.barrier()
            #qc.x(address)
        if m == problem_set[2]:
            qc.x(address[0:2])
            qc.x(address[3])
            for i in m:
                qc.mct([address[0],address[1],address[2],address[3],row[int(i[0])]],column[int(i[1])],mode = 'noancilla') 
            qc.x(address[0:2])
            qc.x(address[3])
            qc.barrier()
            #qc.x(address)
        if m == problem_set[3]:
            qc.x(address[0:2])
            for i in m:
                qc.mct([address[0],address[1],address[2],address[3],row[int(i[0])]],column[int(i[1])],mode = 'noancilla')
            qc.x(address[0:2])
            qc.barrier()
        if m == problem_set[4]:
            qc.x(address[0])
            qc.x(address[2:4])
            for i in m:
                qc.mct([address[0],address[1],address[2],address[3],row[int(i[0])]],column[int(i[1])],mode = 'noancilla')
            qc.x(address[2:4])
            qc.x(address[0])
            qc.barrier()
        if m == problem_set[5]:
            qc.x(address[0])
            qc.x(address[2])
            for i in m:
                qc.mct([address[0],address[1],address[2],address[3],row[int(i[0])]],column[int(i[1])],mode = 'noancilla')
            qc.x(address[0])
            qc.x(address[2])
            qc.barrier()
        if m == problem_set[6]:
            qc.x(address[1:3])
            for i in m:
                qc.mct([address[0],address[1],address[2],address[3],row[int(i[0])]],column[int(i[1])],mode = 'noancilla')
            qc.x(address[1:3])
            qc.barrier()
        if m == problem_set[7]:
            qc.x(address[0])
            for i in m:
                qc.mct([address[0],address[1],address[2],address[3],row[int(i[0])]],column[int(i[1])],mode = 'noancilla')
            qc.x(address[0])
            qc.barrier()
        if m == problem_set[8]:
            qc.x(address[1:4])
            for i in m:
                qc.mct([address[0],address[1],address[2],address[3],row[int(i[0])]],column[int(i[1])],mode = 'noancilla')
            qc.x(address[1:4])
            qc.barrier()
        if m == problem_set[9]:
            qc.x(address[1:3])
            for i in m:
                qc.mct([address[0],address[1],address[2],address[3],row[int(i[0])]],column[int(i[1])],mode = 'noancilla')
            qc.x(address[1:3])
            qc.barrier()
        if m == problem_set[10]:
            qc.x(address[1])
            qc.x(address[3])
            for i in m:
                qc.mct([address[0],address[1],address[2],address[3],row[int(i[0])]],column[int(i[1])],mode = 'noancilla')
            qc.x(address[3])
            qc.x(address[1])
            qc.barrier()
        if m == problem_set[11]:
            qc.x(address[1])
            for i in m:
                qc.mct([address[0],address[1],address[2],address[3],row[int(i[0])]],column[int(i[1])],mode = 'noancilla')       
            qc.x(address[1])
            qc.barrier()
        if m == problem_set[12]:
            qc.x(address[2:4])
            for i in m:
                qc.mct([address[0],address[1],address[2],address[3],row[int(i[0])]],column[int(i[1])],mode = 'noancilla')
            qc.x(address[2:4])
            qc.barrier()
        if m == problem_set[13]:
            qc.x(address[2])
            for i in m:
                qc.mct([address[0],address[1],address[2],address[3],row[int(i[0])]],column[int(i[1])],mode = 'noancilla')            
            qc.x(address[2])
            qc.barrier()
        if m == problem_set[14]:
            qc.x(address[3])
            for i in m:
                qc.mct([address[0],address[1],address[2],address[3],row[int(i[0])]],column[int(i[1])],mode = 'noancilla')
            qc.x(address[3])
            qc.barrier()
        if m == problem_set[15]:
            #qc.x(address)
            for i in m:
                qc.mct([address[0],address[1],address[2],address[3],row[int(i[0])]],column[int(i[1])],mode = 'noancilla')         
            #qc.x(address)
            qc.barrier()
```

```python
def oracle(qc,row_qubits,column_qubits,search_qubits,flag_qubit):
    for i in range(4):
        qc.cx(row_qubits[i],search_qubits[i])
    qc.barrier()
    for i in range(4):    
        qc.cx(column_qubits[i],search_qubits[i+4])
    qc.barrier()
    qc.mct(search_qubits,flag_qubit)
    qc.barrier()
    for i in range(4):
        qc.cx(row_qubits[i],search_qubits[i])
    qc.barrier()
    for i in range(4):    
        qc.cx(column_qubits[i],search_qubits[i+4])
    qc.barrier()
```

```python
def oracle_dagger(qc,row_qubits,column_qubits,search_qubits,flag_qubit):
    for i in range(4):
        qc.cx(row_qubits[i],search_qubits[i])
    qc.barrier()
    for i in range(4):    
        qc.cx(column_qubits[i],search_qubits[i+4])
    qc.barrier()
    qc.mct(search_qubits,flag_qubit)
    qc.barrier()
    for i in range(4):
        qc.cx(row_qubits[i],search_qubits[i])
    qc.barrier()
    for i in range(4):    
        qc.cx(column_qubits[i],search_qubits[i+4])
    qc = qc.inverse()
    qc.barrier()
```

```python
def diffuser(qc,var_qubits):
    qc.h(var_qubits[:])
    qc.x(var_qubits[:])
    qc.h(var_qubits[7])
    qc.mct(var_qubits[0:7], var_qubits[7])
    qc.h(var_qubits[7])
    qc.x(var_qubits[:])
    qc.h(var_qubits[:])
```

```python
def counter(qc,search,auxiliary):
    for i in range(len(search)):
        qc.mct([search[i],auxiliary[0],auxiliary[1],auxiliary[2]],auxiliary[3],mode = 'noancilla')
        qc.mct([search[i],auxiliary[0],auxiliary[1]],auxiliary[2],mode = 'noancilla')
        qc.ccx(search[i],auxiliary[0],auxiliary[1])
        qc.cx(search[i],auxiliary[0])
    qc.barrier()

def counter_dagger(qc,search,auxiliary):
    for i in range(len(search)):
        qc.mct([search[i],auxiliary[0],auxiliary[1],auxiliary[2]],auxiliary[3],mode = 'noancilla')
        qc.mct([search[i],auxiliary[0],auxiliary[1]],auxiliary[2],mode = 'noancilla')
        qc.ccx(search[i],auxiliary[0],auxiliary[1])
        qc.cx(search[i],auxiliary[0])
    qc = qc.inverse()
    qc.barrier()
```

```python
qram(qc,address_qubits,row_qubits,column_qubits,problem_set)
oracle(qc,row_qubits,column_qubits,search_qubits,flag_qubit)
diffuser(qc,search_qubits)
qc.barrier()
counter(qc,search_qubits,counter_qubits)
qc.ccx(counter_qubits[2],counter_qubits[3],flag_qubit)
qc.barrier()
counter_dagger(qc,search_qubits,counter_qubits)
diffuser(qc,search_qubits)
qc.barrier()
oracle_dagger(qc,row_qubits,column_qubits,search_qubits,flag_qubit)
qram(qc,address_qubits,row_qubits,column_qubits,problem_set)
qc.h(address_qubits[:])
qc.x(address_qubits[:])
qc.h(address_qubits[3])
qc.mct(address_qubits[0:3], address_qubits[3])
qc.h(address_qubits[3])
qc.x(address_qubits[:])
qc.h(address_qubits[:])
qc.barrier()
qc.measure([3,2,1,0],classical_bits)
qc.draw('mpl')
```
